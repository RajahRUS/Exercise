```python
# Dictionaries

raw_data = [
    {"user_id": "user_1", "amount": 100},
    {"user_id": "user_2", "amount": 200},
    {"user_id": "user_1", "amount": 150},
]
```

```python
user_spending = {}
for record in raw_data:
    uid = record["user_id"]
    amt = record["amount"]
    # Метод .get(key, default) позволяет избежать ошибки, если ключа еще нет
    user_spending[uid] = user_spending.get(uid, 0) + amt
```

```python
user_spending
```

```python
# Tuples

orders = [
    ("Moscow", "Electronics"),
    ("SPB", "Food"),
    ("Moscow", "Electronics"),
    ("Moscow", "Food"),
]
```

```python
stats = {}
for city, category in orders:
    key = (city, category) # Кортеж как ключ
    stats[key] = stats.get(key, 0) + 1
```

```python
print(stats)
```

```python
users_data = [
    {"name": "Alice", "buys": [1, 4]},
    {"name": "Bob", "buys": [5]},
    {"name": "Charlie", "buys": [6] }
]
```

```python
users_data
```

```python
results = []
for user in users_data:
    avg_bill = sum(user["buys"]) / len(user["buys"])
    results.append((user["name"], avg_bill)) # Сохраняем как список кортежей

results
```

```python
print(results.sort(key=lambda x: x[1], reverse=True))
```

```python
print(f"Top user: {results}, Average bill: {results[2]}")
```
