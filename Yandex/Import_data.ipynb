```python
import xml.etree.ElementTree as ET
import pandas as pd
import requests
```

```python
url = 'https://www.w3schools.com/xml/cd_catalog.xml'
response = requests.get(url)

# 2. Загрузка и парсинг из строки
root = ET.fromstring(response.content)
```

```python
root# 3. Преобразование в список словарей
# В этом XML каждый объект находится в теге <CD>
data = []
for item in root.findall('.//CD'): 
    # Создаем словарь для каждой строки (ID, Artist, Company и т.д.)
    row = {child.tag: child.text for child in item}
    data.append(row)
```

```python
# 4. Перевод в DataFrame для проверки гипотез [3]
df_from_tree = pd.DataFrame(data)
```

```python
print("Данные, полученные через ElementTree (список словарей):")
print(df_from_tree.head())
```
